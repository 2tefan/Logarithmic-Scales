## Setup âš™ï¸

```python
# Import matplotlib & numpy
import matplotlib.pyplot as plt
import numpy as np

# Enable inline for notebook
%matplotlib inline
```

```python
# Define function to plot graphs
def plot(x: np.ndarray, y: np.ndarray or list, title: str, labels: str = None) -> None:
    figure, axis = plt.subplots(1, 3, figsize=(25, 7))
    figure.suptitle(title, fontsize=39)
    figure.supxlabel("X")
    figure.supylabel("Y")

    for a in axis:
        a.grid()
        if labels is None:
            a.plot(x, y, "r")
        else:
            for line, label in zip(y, labels):
                a.plot(x, line, label=label)
                a.legend()

    axis[1].set_yscale("log")

    axis[2].set_yscale("log")
    axis[2].set_xscale("log")
```

## Plotting ðŸ“ˆ

```python
x = np.logspace(start=0.01, stop=1, num=1000)
```

```python
y = x
plot(x, y, "y(x)=x")
```

```python
y = x**2
plot(x, y, "y(x)=xÂ²")
```

```python
y = 2**x
plot(x, y, "y(x)=2^x")
```

## Variations ðŸ§®

```python
y1 = 1 * x
y2 = 2000 * x
y3 = 5000 * x
plot(x, [y1, y2, y3], "y(x)=k*x", ["k=1", "k=2000", "k=5000"])
```

```python
y1 = x
y2 = x + 20
y3 = x + 50
plot(x, [y1, y2, y3], "y(x)=x+d", ["d=0", "d=20", "d=50"])
```

```python
y1 = x**2
y2 = x**20
y3 = x**50
plot(x, [y1, y2, y3], "y(x)=x^c", ["c=2", "c=20", "c=50"])
```

```python
y1 = 2**x
y2 = 20**x
y3 = 50**x
plot(x, [y1, y2, y3], "y(x)=c^x", ["c=2", "c=20", "c=50"])
```
